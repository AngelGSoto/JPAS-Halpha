```python
from astropy.io import fits
import numpy as np
from astropy.table import Table
import pandas as pd
from astropy.table import vstack
from matplotlib import pyplot as plt
import pandas as pd
import glob
```

```python
ruta = "../Halpha_emitters" 
```

```python
archivos_csv = glob.glob(ruta + "/*.csv")
```

```python
# Lee cada archivo CSV y almacena el DataFrame en una lista
lista_dataframes = [pd.read_csv(archivo) for archivo in archivos_csv]
```

```python
# Concatena todos los DataFrames en uno solo
df_concatenado = pd.concat(lista_dataframes, ignore_index=True)
```

```python
df_concatenado
```

```python
# Obtener los nombres de las columnas
column_names = df_concatenado.columns.tolist()
# Mostrar los nombres de las columnas
print("Nombres de las columnas:", column_names)
```

```python
plt.scatter(df_concatenado["color_x"], df_concatenado["color_y"], marker=".", edgecolors="k")
# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# Show the plot
plt.show()
```
